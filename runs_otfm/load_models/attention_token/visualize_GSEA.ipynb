```python
import pandas as pd
import seaborn as sns
```

```python
df_1 = pd.read_csv('spearman_corrs_1.csv')
df_2 = pd.read_csv('spearman_corrs_2.csv')
df_3 = pd.read_csv('spearman_corrs_3.csv')
df_4 = pd.read_csv('spearman_corrs_4.csv')
```

```python
d_1 = pd.read_csv('/home/icb/dominik.klein/git_repos/ot_pert_new/competing_methods/combosciplex/additive_model/add_spearman_corrs_1.csv')
d_2 = pd.read_csv('/home/icb/dominik.klein/git_repos/ot_pert_new/competing_methods/combosciplex/additive_model/add_spearman_corrs_2.csv')
d_3 = pd.read_csv('/home/icb/dominik.klein/git_repos/ot_pert_new/competing_methods/combosciplex/additive_model/add_spearman_corrs_3.csv')
d_4 = pd.read_csv('/home/icb/dominik.klein/git_repos/ot_pert_new/competing_methods/combosciplex/additive_model/add_spearman_corrs_4.csv')
```

```python
to_remove = ["REACTOME_CELL_CYCLE_MITOTIC", "REACTOME_CLASS_I_MHC_MEDIATED_ANTIGEN_PROCESSING_PRESENTATION"]
```

```python
df_1 = df_1[~df_1["Unnamed: 0"].isin(to_remove)]
df_2 = df_2[~df_2["Unnamed: 0"].isin(to_remove)]
df_3 = df_3[~df_3["Unnamed: 0"].isin(to_remove)]
df_4 = df_4[~df_4["Unnamed: 0"].isin(to_remove)]

d_1 = d_1[~d_1["Unnamed: 0"].isin(to_remove)]
d_2 = d_2[~d_2["Unnamed: 0"].isin(to_remove)]
d_3 = d_3[~d_3["Unnamed: 0"].isin(to_remove)]
d_4 = d_4[~d_4["Unnamed: 0"].isin(to_remove)]
```

```python
df_1["model"] = "CellFlow"
df_2["model"] = "CellFlow"
df_3["model"] = "CellFlow"
df_4["model"] = "CellFlow"

d_1["model"] = "additive"
d_2["model"] = "additive"
d_3["model"] = "additive"
d_4["model"] = "additive"
```

```python
dff_1 = pd.concat((df_1, d_1))
```

```python
ax = sns.boxplot(dff_1, x="model", y="spearman_correlation")
ax.set_ylim([-1, 1]) 
```

```python
dff_2 = pd.concat((df_2, d_2))

ax = sns.boxplot(dff_2, x="model", y="spearman_correlation")
ax.set_ylim([-1, 1]) 
```

```python
dff_3 = pd.concat((df_3, d_3))

ax = sns.boxplot(dff_3, x="model", y="spearman_correlation")
ax.set_ylim([-1, 1]) 
```

```python
dff_4 = pd.concat((df_4, d_4))

ax = sns.boxplot(dff_4, x="model", y="spearman_correlation")
ax.set_ylim([-1, 1]) 
```

```python
df_all = pd.concat((df_1, df_2, df_3, df_4, d_1, d_2, d_3, d_4))
```

```python
import matplotlib.pyplot as plt
ax = sns.boxplot(df_all, x="Unnamed: 0", y="spearman_correlation", hue="model")
plt.xticks(rotation=90)
```
