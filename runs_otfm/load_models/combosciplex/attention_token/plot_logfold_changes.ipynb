```r
library(tidyverse)
```

```r
log_fold_changes <- read_csv("/lustre/groups/ml01/workspace/ot_perturbation/analysis/combosciplex/logfoldchanges_split_3.csv") 
```

```r
unique_vals <- unique(log_fold_changes$condition)
```

```r
df1 <- log_fold_changes %>%
  filter(condition == unique_vals[[1]])

model <- lm(log_fold_change_pred ~ log_fold_change, data = df1)
r_squared <- summary(model)$r.squared

ggplot(df1, aes(x = log_fold_change, y = log_fold_change_pred, color = is_deg)) +
  geom_point(size = 5, alpha = 0.7) +  # Adjusting point size and adding transparency
  scale_color_manual(values = c("TRUE" = "blue", "FALSE" = "gray")) +
  labs(x = "Log Fold Change", 
       y = "Predicted Log Fold Change", 
       color = "Differentially Expressed") +  # Adding better labels
  theme_minimal(base_size = 14) +  
  geom_abline(slope = 1, intercept = 0, color="red", size=2) +
  xlim(-1.5, 1.5) +  # Setting x-axis limits
  ylim(-1.5, 1.5) +  # Setting y-axis limits
  ggtitle(unique_vals[[1]]) + 
  theme(
    panel.grid.major = element_line(color = "lightgray", size = 0.5),  # Making grid lines subtler
    panel.grid.minor = element_blank(),  # Removing minor grid lines for a cleaner look
    axis.line = element_line(color = "black"),  # Adding axis lines
    legend.position = "top",  # Placing the legend at the top for better visibility
    legend.title = element_text(size = 12),  # Adjusting legend title size
    legend.text = element_text(size = 10)    # Adjusting legend text size
  )+
  annotate("text", x = 1.3, y = -1.4, label = paste("R² =", round(r_squared, 3)), 
           size = 15, color = "black", hjust = 1)  # Adding the R-squared annotation at bottom right

```
