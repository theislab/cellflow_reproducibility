```python
import pandas as pd
def gmt_to_decoupler(pth: Path) -> pd.DataFrame:
    """
    Parse a gmt file to a decoupler pathway dataframe.
    """
    from itertools import chain, repeat

    pathways = {}

    with Path(pth).open("r") as f:
        for line in f:
            name, _, *genes = line.strip().split("\t")
            pathways[name] = genes

    return pd.DataFrame.from_records(
        chain.from_iterable(zip(repeat(k), v) for k, v in pathways.items()),
        columns=["geneset", "genesymbol"],
    )

```

```python
from typing import Any, Literal
from sklearn.decomposition import KernelPCA
from sklearn.cross_decomposition import CCA

def predict_with_cca(embeddings_seen: pd.DataFrame, target_variable: pd.Series, embeddings_unseen: pd.DataFrame, kernel_pca_mode: Literal["linear", "poly", "rbf", "sigmoid", "cosine"] = "cosine", kernel_pca_n_components: int = 10) -> pd.Series:
    """Predict target variable for unseen data using canonical correlation analysis (CCA).

    Parameters
    ----------
    embeddings_seen
        Embeddings of the seen data. Index corresponding to condition names, values to embeddings.
    target_variable
        Target variable of the seen data. Index corresponding to condition names, values to target variable.
    embeddings_unseen
        Embeddings of the unseen data. Index corresponding to condition names, values to embeddings which to 
        predict the target variable.
    kernel_pca_mode
        Kernel for the KernelPCA.
    kernel_pca_n_components
        Number of components to keep in the KernelPCA.
    

    Returns
    -------
    Predicted target variable for the unseen data.
    """
    kpca = KernelPCA(n_components=kernel_pca_n_components, kernel=kernel_pca_mode)
    X = embeddings_seen.values
    y = target_variable.loc[embeddings_seen.index].values.astype("float64")

    X_transformed = kpca.fit_transform(
        X
    )

    cca = CCA(n_components=1) # TODO: possibly extend to multiple variables
    cca.fit(X_transformed, y)
    _, y_c = cca.transform(X_transformed, y)
    print(y_c.squeeze().shape, y.shape)
    correct_orientation = np.corrcoef(y_c.squeeze(), y)[0,1] > 0.0
    
    
    X_new = embeddings_unseen.values
    X_new_transformed = kpca.transform(X_new)
    y_pred = cca.predict(X_new_transformed)

    return pd.Series(index=embeddings_unseen.index, data=y_pred * (1.0 if correct_orientation else -1.0))
```

```python
from typing import Any, Literal
from sklearn.decomposition import KernelPCA
from sklearn.cross_decomposition import CCA
from cca_zoo.nonparametric import KCCA

def predict_with_kernel_cca(embeddings_seen: pd.DataFrame, target_variable: pd.Series, embeddings_unseen: pd.DataFrame, kernel: Literal["linear", "poly", "rbf", "sigmoid", "cosine"] = "cosine", kernel_pca_n_components: int = 10) -> pd.Series:
    """Predict target variable for unseen data using canonical correlation analysis (CCA).

    Parameters
    ----------
    embeddings_seen
        Embeddings of the seen data. Index corresponding to condition names, values to embeddings.
    target_variable
        Target variable of the seen data. Index corresponding to condition names, values to target variable.
    embeddings_unseen
        Embeddings of the unseen data. Index corresponding to condition names, values to embeddings which to 
        predict the target variable.
    kernel_pca_mode
        Kernel for the KernelPCA.
    kernel_pca_n_components
        Number of components to keep in the KernelPCA.
    

    Returns
    -------
    Predicted target variable for the unseen data.
    """
    X = embeddings_seen.values
    X_mean = X.mean(axis=0)
    X -= X_mean
    y = target_variable.loc[embeddings_seen.index].values.reshape(-1, 1).astype('float64')
    y_mean = y.mean(axis=0)
    y -= y.mean(axis=0)

    
    kcca = KCCA(latent_dimensions=1, kernel=kernel) # TODO: possibly extend to multiple variables
    kcca.fit((X, y))
    _, y_c = kcca.transform((X, y))
    print(y_c.squeeze().shape, y.squeeze().shape)
    correct_orientation = np.corrcoef((y_c.squeeze()), y.squeeze())[0,1] > 0.0
    
    
    X_new = embeddings_unseen.values
    X -= X_mean
    y_pred = kcca.transform((X_new, None))[0]

    return pd.Series(index=embeddings_unseen.index, data=y_pred.squeeze() * (1.0 if correct_orientation else -1.0))
```

```python
pathway_dfs = {}

for pathway in pathways:
    dict_gsea_results = {}

    gene_set = reactome[reactome.geneset==pathway]
    
    for condition, t_stats in dict_t_stats.items():
        scores, norm, pvals = decoupler.run_gsea(
            t_stats.T,
            gene_set,
            source="geneset",
            target="genesymbol",
            min_n=1
        )
        gsea_results = (
            pd.concat({"score": scores.T, "norm": norm.T, "pval": pvals.T}, axis=1)
            .droplevel(level=1, axis=1)
            .sort_values("pval")
        )
        dict_gsea_results[condition] = gsea_results
    res_score = {}

    for cond, df in dict_gsea_results.items():
        res_score[cond] = df.loc[pathway]["score"]
    df = pd.DataFrame(res_score, index=[0]).T
    df["condition"] = df.index
    df["mode"] = df.apply(get_mode, axis=1)
    df["score"] = df[0]
    pathway_dfs[pathway] = df
```

```python
res = {}
for kernel in ['sigmoid',  'poly', 'laplacian', 'cosine', 'rbf', 'linear', 'polynomial']:
    corr_result = {}
    
    for pathway, df in pathway_dfs.items():
        df_seen = df[df["mode"]=="seen"]
        df_unseen = df[df["mode"]=="true"]
        y = df_seen["score"]
        df_emb_ordered = df_emb.loc[df_seen.index]
        out = predict_with_kernel_cca(df_emb.loc[df_seen.index], y, df_emb.loc[df_unseen.index], kernel=kernel, kernel_pca_n_components=5)
        df_unseen["cca_score"] = out
        df_unseen["score_true_rank"] = df_unseen["score"].rank()
        df_unseen["cca_score_rank"] = df_unseen["cca_score"].rank()
        corr_result[pathway] = df_unseen[["score_true_rank", "cca_score"]].corr(method="spearman").iloc[0,1]
    res[kernel]= np.array([v for v in corr_result.values()]).mean()
```
