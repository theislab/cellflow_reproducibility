```python
import scanpy as sc
import anndata as ad
```

```python
split = 1
```

```python
adata_train_path = f"/lustre/groups/ml01/workspace/ot_perturbation/data/combosciplex/adata_train_{split}.h5ad"
adata_test_path = f"/lustre/groups/ml01/workspace/ot_perturbation/data/combosciplex/adata_test_{split}.h5ad"
adata_ood_path = f"/lustre/groups/ml01/workspace/ot_perturbation/data/combosciplex/adata_ood_{split}.h5ad"
adata_train = sc.read_h5ad(adata_train_path)
adata_test = sc.read_h5ad(adata_test_path)
adata_ood = sc.read_h5ad(adata_ood_path)
    
```

```python
adata_full = ad.concat((adata_train, adata_test, adata_ood))
```

```python
sc.pl.umap(adata_full, color="condition")
```

```python
sc.pl.umap(adata_full, color="condition", size=10, title="Perturbation")
```

```python
palette = {}
palette[True] =  "#1099c7"
palette[False] = "#d3d3d3"
        
```

```python
import numpy as np
category_indices = adata_full.obs['control'] == True
sizes = np.where(category_indices, 50, 10)
sc.pl.umap(adata_full, color="control", palette=palette, size=sizes)
```

```python
import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt

# Define the category you want to color differently
category = "not ood"

# Get the index of the points corresponding to "not ood"
category_indices = adata_full.obs['ood_1'] == category

# Create a size array where "not ood" has smaller markers, rest have larger ones
sizes = np.where(category_indices, 5, 20)

# Define the custom color map, setting "not ood" to #d3d3d3
# Get unique categories from 'ood_1'
unique_categories = adata_full.obs['ood_1'].unique()

# Create a color palette with random colors for categories except "not ood"
palette = {cat: plt.cm.tab20(i) for i, cat in enumerate(unique_categories) if cat != "not ood"}

# Add the specific color for "not ood"
palette["not ood"] = "#d3d3d3"

# Plot with the custom colors and sizes
sc.pl.umap(adata_full, color="ood_1", palette=palette, size=sizes)
```
