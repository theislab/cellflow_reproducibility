```r
library(Seurat)
library(Signac)
library(anndata)
library(readr)
library(Matrix)
```

```r
path <- "/lustre/groups/ml01/workspace/ot_perturbation/data/multiome/GLI3_KO_multiome_18d_srt.rds"
```

```r
data <- readRDS(path)
```

```r
data
```

```r
DimPlot(data, reduction = "umap")
```

```r
adata <- anndata::AnnData(X = t(GetAssayData(data, slot = "counts")))
```

```r
adata$obs = data@meta.data
```

```r
adata$write_h5ad("/lustre/groups/ml01/workspace/ot_perturbation/data/multiome/GLI3_KO_multiome_18d_srt_rna.h5ad",
                compression = "gzip")
```
