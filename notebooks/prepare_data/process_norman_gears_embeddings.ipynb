```python
import scanpy as sc
import pickle
import os
import anndata
import numpy as np
import pandas as pd

output_dir = "/lustre/groups/ml01/workspace/ot_perturbation/data/norman"
```

```python
go_path = "/lustre/groups/ml01/workspace/ot_perturbation/pert_embeddings/norman/gears/data/norman/go.csv"
gene_path = "/lustre/groups/ml01/workspace/ot_perturbation/pert_embeddings/norman/gears/data/essential_all_data_pert_genes.pkl"
df = pd.read_csv(go_path)
df = df.groupby('target').apply(lambda x: x.nlargest(20 + 1,['importance'])).reset_index(drop = True)
with open(gene_path, 'rb') as f:
    gene_list = pickle.load(f)
    
df = df[df["source"].isin(gene_list)]
```

```python
df
```

```python
def get_map(pert):
    tmp = pd.DataFrame(np.zeros(len(gene_list)), index=gene_list)
    tmp.loc[df[df.target == pert].source.values, :] = df[df.target == pert].importance.values[:, np.newaxis]
    return tmp.values.flatten()    
    
```

Train contains all the genes already.

```python
train = sc.read_h5ad(os.path.join(output_dir, "adata_train_2_seen_genes.h5ad"))
ood = sc.read_h5ad(os.path.join(output_dir, "adata_ood_2_seen_genes.h5ad"))
test = sc.read_h5ad(os.path.join(output_dir, "adata_test_2_seen_genes.h5ad"))
```

```python
train.obs
```

```python
pert2neighbor =  {i: get_map(i) for i in list(train.obs["gene_1"].cat.categories)}    
#adata.uns["pert2neighbor"] = pert2neighbor
```

```python
pert2neighbor_array = np.asarray([val for val in pert2neighbor.values()])
keep_idx = pert2neighbor_array.sum(0) > 0
```

```python
pert2neighbor_array[:, keep_idx].shape
```

```python
pert2neighbor_array = pert2neighbor_array[:, keep_idx]
```

```python
pert2neighbor_dict = {key: array for key, array in zip(pert2neighbor.keys(), pert2neighbor_array)}   
```

```python
train.uns["pert2neighbor"] = pert2neighbor
ood.uns["pert2neighbor"] = pert2neighbor
test.uns["pert2neighbor"] = pert2neighbor
```

```python
train.uns["pert2neighbor_prepared"] = pert2neighbor_dict
ood.uns["pert2neighbor_prepared"] = pert2neighbor_dict
test.uns["pert2neighbor_prepared"] = pert2neighbor_dict
```

```python
embedding_dim = pert2neighbor_array.shape[1]
```

```python
def get_embedding(adata, pert_dict, embedding_dim):
    
    emb_1_list = []
    emb_2_list = []
    
    # Iterate through gene_1 and gene_2 columns
    for gene_1, gene_2 in zip(adata.obs['gene_1'], adata.obs['gene_2']):
        
        # Fetch embeddings from embedding_dict
        emb_1 = pert_dict.get(gene_1, np.zeros(embedding_dim))
        emb_2 = pert_dict.get(gene_2, np.zeros(embedding_dim))
        
        emb_1_list.append(emb_1.astype(np.float32))
        emb_2_list.append(emb_2.astype(np.float32))
    
    emb_1_array = np.array(emb_1_list)
    emb_2_array = np.array(emb_2_list)
    
    adata.obsm['go_emb_1'] = emb_1_array
    adata.obsm['go_emb_2'] = emb_2_array

    return adata
```

```python
train = get_embedding(train, pert2neighbor_dict, embedding_dim)
test = get_embedding(test, pert2neighbor_dict, embedding_dim)
ood = get_embedding(ood, pert2neighbor_dict, embedding_dim)
```

```python
train.write(os.path.join(output_dir, "adata_train_2_seen_genes.h5ad"))
ood.write(os.path.join(output_dir, "adata_ood_2_seen_genes.h5ad"))
test.write(os.path.join(output_dir, "adata_test_2_seen_genes.h5ad"))
```
