```python
def mapping(label):
    if label == 'ctrl-inj':
        ret = 'control'
    elif 'ctrl' in label:
        ret = 'control'
    elif 'mut' in label or 'remove' in label:
        ret = label.split('-')[0]
    elif '-' in label:
        return label.split('-')[0], label.split('-')[1]
    else:
        return label, 'negative'
    return ret, 'negative'
```

```python
adata.obs['gene1+gene2'] = [f'{mapping(x)[0]}+{mapping(x)[1]}' for x in adata.obs.gene_target]
```

```python
adata.obs[['gene_target_1', 'gene_target_2']] = adata.obs['gene1+gene2'].str.split('+', expand=True)
```

```python
adata.obs["gene_target_1"] = adata.obs.apply(lambda x: "control" if x['gene_target_1']=="negative" else x['gene_target_1'], axis=1)
adata.obs["gene_target_2"] = adata.obs.apply(lambda x: "control" if x['gene_target_2']=="negative" else x['gene_target_2'], axis=1)
```

```python
sc.pp.normalize_total(adata)
sc.pp.log1p(adata)
```

```python
sc.pp.highly_variable_genes(adata, inplace=True, n_top_genes=2000)
adata = adata[:,adata.var["highly_variable"]==True]
```

```python
adata.obs["condition"] = adata.obs.apply(lambda x: x["gene_target_1"]+"_"+x["gene_target_2"]+"_"+x["timepoint"], axis=1)
```

```python
adata.obs["condition"].value_counts().values.min(), len(adata.obs["condition"].value_counts())
```

```python
adata.obs["is_control"] = adata.obs.apply(lambda x: True if (x["gene_target_1"]=="control") and (x["gene_target_2"]=="control") else False, axis=1)
```

```python
adata.obs["is_control"].value_counts()
```

```python
condition_counts = adata.obs.groupby('timepoint')['gene1+gene2'].nunique()
unique_conditions = adata.obs.groupby('timepoint')['gene1+gene2'].apply(pd.unique).reset_index()
unique_conditions['n_conditions'] = condition_counts.values
unique_conditions
```

```python
import pickle
import os
with open(os.path.join("/lustre/groups/ml01/workspace/ot_perturbation/data/zebrafish/ESM2_embeddings.pkl"), 'rb') as file:
    data = pickle.load(file)
```

```python
unique_gene_targets = set(adata.obs["gene_target_1"].unique()).union(set((adata.obs["gene_target_1"].unique())))
```

```python
gene_embeddings = {k: data[k] for k in unique_gene_targets if k!="control"}
gene_embeddings["control"] = data["negative"]
adata.uns["gene_embeddings"] = gene_embeddings
```

```python
adata.obs["timepoint"] = adata.obs["timepoint"].astype(float)
```

```python
import numpy as np
adata.obs["logtime"] = adata.obs["timepoint"].apply(np.log)
```

```python
adata.write("/lustre/groups/ml01/workspace/ot_perturbation/data/zebrafish_new/zebrafish_processed.h5ad")
```
