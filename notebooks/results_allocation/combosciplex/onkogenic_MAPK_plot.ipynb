```r
# Create a new column 'hjust_label' for conditional horizontal justification
df_res <- df_res %>%
  mutate(hjust_label = ifelse(row_number() <= 4, -0.0, 1.0))

# Now use this new column inside geom_text()
p <- ggplot(df_res, aes(x=rank_true, y=rank_pred, color=condition)) +
  geom_point() +  # Scatter plot
  geom_text(aes(
    label = condition, 
    hjust = hjust_label  # Use the new column for hjust
  ), vjust = -0.5, size = 5) +  # Position vertically slightly above
  labs(
    x = "True rank among OOD drugs of pathway activation",
    y = "Predicted rank among OOD\n drugs of pathway activation"
  ) +
  theme_bw() +
  scale_color_manual(values = combosciplex_color_dict) +
  theme(
    axis.title.x = element_text(size=16, face="bold"),
    axis.title.y = element_text(size=16, face="bold"),
    axis.text.x = element_text(size=14),
    axis.text.y = element_text(size=14),
    legend.title = element_text(size=14),
    legend.text = element_text(size=12),
    legend.position = "right",
    legend.box.background = element_rect(color="black"),
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank()
  )# Apply a minimal theme

p
```
