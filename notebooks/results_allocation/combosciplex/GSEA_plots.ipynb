```r
library(tidyverse)
```

```r
color_dict = c("genot"= "#317d01",
    "CellFlow"= "#59d1d6",
    "chemCPA"= "#e6b1fc",
    "CPA"= "#6c0399",
    "Baseline"="#8f97a8",
    "Additive"="#ffA500")
```

```r
output_dir <- "/lustre/groups/ml01/workspace/ot_perturbation/data/combosciplex/results/plots"
```

# Decoded

```r
df_1 <- read_csv("/lustre/groups/ml01/workspace/ot_perturbation/data/combosciplex/results/allocation/GSEA_results_1.csv") 

```

```r
options(repr.plot.width = 12, repr.plot.height = 6)
df_1$model <- factor(df_1$model, 
                                levels = c("additive", "CPA", "chemCPA", "CellFlow"),
                                labels = c("Additive", "CPA", "chemCPA", "CellFlow"))  # Rename levels

p <- ggplot(df_1, aes(x=model, y=spearman_correlation, fill=model)) +
  geom_boxplot(outlier.size=2) +
  labs(x="Model", y="Spearman correlation GSEA score") +
  scale_fill_manual(values=color_dict, labels=c("Additive", "CPA", "chemCPA", "CellFlow")) +
  scale_x_discrete(labels=c("")) +  # Change x-axis labels
  theme_bw() +
  ylim(-1,1) + 
  theme(
    axis.title.x = element_text(size=16, face="bold"),
    axis.title.y = element_text(size=16, face="bold"),
    axis.text.x = element_text(size=14),
    axis.text.y = element_text(size=14),
    legend.title = element_text(size=14),
    legend.text = element_text(size=12),
    legend.position = "right",
    legend.box.background = element_rect(color="black"),
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank()
  )

p
```

```r
ggsave(file.path(output_dir, "gsea_allocated_1.pdf"), plot = p, device = "pdf")
```
