```python
def get_perturbation_effects(model, ood_cell_type, ood_subset, seen_subset, cell_types, knockouts, pathway, gene_indices):
    targets = {}
    sources = {}
    for cell_type in cell_types:
        for knockout in knockouts:
                if cell_type == ood_cell_type:
                    ood_source = ood_subset[ood_subset.obs['condition'] == f"{cell_type}_{pathway}_NT"]
                    ood_target = ood_subset[ood_subset.obs['condition'] == f"{cell_type}_{pathway}_{knockout}"]

                    recon_source = reconstruct_data(ood_source.obsm['X_pca'])[:, gene_indices]
                    recon_target = reconstruct_data(ood_target.obsm['X_pca'])[:, gene_indices]

                    targets[(cell_type + '-gt', pathway, knockout)] = recon_target
                    sources[(cell_type + '-gt', pathway, knockout)] = recon_source

                    covariate_data = ood_target.obs[['cell_type', 'gene', 'pathway']]
                    pred_data = model.predict(ood_source, sample_rep='X_pca', covariate_data=covariate_data)
                    pred_data = reconstruct_data(pred_data[(knockout, pathway, cell_type)])[:, gene_indices]

                    targets[(cell_type, pathway, knockout)] = pred_data
                    sources[(cell_type, pathway, knockout)] = recon_source
                    
                else: 
                    recon_source = reconstruct_data(seen_subset[seen_subset.obs['condition'] == f"{cell_type}_{pathway}_NT"].obsm['X_pca'])[:, gene_indices]
                    recon_target = reconstruct_data(seen_subset[seen_subset.obs['condition'] == f"{cell_type}_{pathway}_{knockout}"].obsm['X_pca'])[:, gene_indices]

                    targets[(cell_type, pathway, knockout)] = recon_target
                    sources[(cell_type, pathway, knockout)] = recon_source

                    # targets[(cell_type, pathway, knockout)] = seen_subset[seen_subset.obs['condition'] == f"{cell_type}_{pathway}_{knockout}"].X.toarray()
                    # sources[(cell_type, pathway, knockout)] = seen_subset[seen_subset.obs['condition'] == f"{cell_type}_{pathway}_NT"].X.toarray()

    return jtu.tree_map(calc_zscores, targets, sources)
```

## Read data once

```python
data_path = cfg.dataset.dataset_path
file_suffix = cfg.dataset.file_suffix

dataset_path = {
    "train": os.path.join(data_path, f"adata_train{file_suffix}.h5ad"),
    "test": os.path.join(data_path, f"adata_test{file_suffix}.h5ad"),
    "ood": os.path.join(data_path, f"adata_ood{file_suffix}.h5ad")
}
adata_ood = sc.read_h5ad(dataset_path["ood"])
adata_ood = adata_ood[adata_ood.obs['pathway'] == ood_cytokine]

adata_test = sc.read_h5ad(dataset_path["test"])
adata_test = adata_test[adata_test.obs['pathway'] == ood_cytokine]
adata_train = sc.read_h5ad(dataset_path["train"])
adata_train = adata_train[adata_train.obs['pathway'] == ood_cytokine]
adata_seen = ad.concat([adata_train, adata_test])

model_path = os.path.join(cfg.logger.save_model_path, name_of_stored_run)
model_checkpoint = get_highest_checkpoint_file(model_path, cfg.logger.project)

def get_mask(x, y, var_names):
    return x[:, [gene in y for gene in var_names]]

#comp_metrics_fn = compute_metrics_fast if fast_metrics else compute_metrics
mask_fn = partial(get_mask, var_names=adata_train.var_names)

reconstruct_data = partial(reconstruct_data_fn, pcs=adata_train.varm['PCs'], means=adata_train.varm['X_mean'])

cf = cfp.model.CellFlow.load(model_checkpoint)

cf.solver.is_trained = True
```

```python
knockouts = ['IFNGR2', 'JAK1', 'JAK2', 'STAT1', 'IFNGR1'] # 'IFNGR1'
pathway = ood_cytokine
cell_types = ['BXPC3', 'HAP1', 'K562', 'A549', 'HT29', 'MCF7']
# genes = ['AHNAK', 'ASTN2']
genes = [
    "THSD4", "FAT1", "HSPG2", "VCL", "MAGI1", "EFNA5", 
    "TPM1", "ASTN2", "ROR1", "AHNAK"
]

pathway = 'IFNG'

gene_indices = [adata_train.var_names.get_loc(gene) for gene in genes if gene in adata_train.var_names]
ood_subset = adata_ood[:, gene_indices]
seen_subset = adata_seen[:, gene_indices]

perturbation_effect_uc1 = get_perturbation_effects(ood_subset, seen_subset, cell_types, knockouts, pathway, gene_indices)
```

```python
df_uc1 = {}

complete_cell_types = np.unique([k[0] for k in perturbation_effect_uc1.keys()])
for knockout in knockouts:
    df_uc1[knockout] = []
    for ct in complete_cell_types:
        pert_per_ct = perturbation_effect_uc1[(ct, pathway, knockout)]
        df_per_ct = pd.DataFrame(np.expand_dims(pert_per_ct, axis=0), columns=genes)
        #df_per_ct = pd.DataFrame(pert_per_ct, columns=gene_names_uc1)
        df_per_ct['cell_type'] = ct
        #df_per_ct = df_per_ct.groupby('cell_type').mean()
        df_per_ct.set_index('cell_type', inplace=True)
        df_per_ct.index.name = None    
        df_uc1[knockout].append(df_per_ct)
    df_uc1[knockout] = pd.concat(df_uc1[knockout])
    df_uc1[knockout] = df_uc1[knockout].reindex(cell_type_order)
```

```python
for knockout in knockouts:
    sns.heatmap(df_uc1[knockout], vmin=vmin, vmax=vmax, cmap=cmap, annot=True)
    plt.title(knockout)
    plt.show()
```
