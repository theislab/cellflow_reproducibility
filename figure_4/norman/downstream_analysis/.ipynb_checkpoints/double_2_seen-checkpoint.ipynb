```r
library(tidyverse)
library(glue)
```

```r
color_dict = c(
    "CellFlow"= "#B12F8C",
    "GEARS"= "#A2E5E7",
    "Biolord"= "#BD6DED",
    "Baseline"="#E1C8AC")
```

```r
split_to_filter <- "double_seen_1"
```

```r
output_dir <- glue("/lustre/groups/ml01/workspace/ot_perturbation/figures/norman/benchmark/plots_{split_to_filter}")
dir.create(output_dir)
```

```r
output_dir
```

# Decoded

```r
results_decoded <- read_csv("/lustre/groups/ml01/workspace/ot_perturbation/data/norman_2/results/allocation/results_norman_prelim2.csv") 
results_decoded <- results_decoded %>%
  filter(subgroup == split_to_filter)
results_decoded$model <- factor(results_decoded$model, 
                                levels = c("identity", "biolord", "gears", "cellflow"),
                                labels = c("Baseline", "Biolord", "GEARS" , "CellFlow"))  # Correct factorization

```

```r

p <- ggplot(results_decoded, aes(x=model, y=decoded_ood_r_squared, fill=model)) +
  geom_boxplot(outlier.size=2) +
  labs(x="Model", y="R squared normalized gene expression") +
  scale_fill_manual(values=color_dict) +  # Removed labels argument
  scale_x_discrete(labels=c("")) +  # Keep this if you want no x-axis labels
  theme_bw() +
  theme(
    axis.title.x = element_text(size=16, face="bold"),
    axis.title.y = element_text(size=16, face="bold"),
    axis.text.x = element_text(size=14),
    axis.text.y = element_text(size=14),
    legend.title = element_text(size=14),
    legend.text = element_text(size=12),
    legend.position = "right",
    legend.box.background = element_rect(color="black"),
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank()
  ) +
  ylim(0.75, NA)  # Ensure the y-axis starts at 0
p
```

```r
ggsave(file.path(output_dir, "ood_r_squared_decoded.pdf"), plot = p, device = "pdf")
```
