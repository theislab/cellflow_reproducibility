```python
import pandas as pd
import os
```

```python
data_dir = "/lustre/groups/ml01/workspace/ot_perturbation/analysis/sciplex"
```

```python
out_dir = '/lustre/groups/ml01/workspace/ot_perturbation/figures/sciplex/pseudodose'
```

```python
df_upregulated = pd.read_csv(os.path.join(data_dir, "upregulated_genes_Dacinostat.csv"))
df_downregulated = pd.read_csv(os.path.join(data_dir, "downregulated_genes_Dacinostat.csv"))
df_upregulation = pd.read_csv(os.path.join(data_dir, "upregulated_data_Dacinostat.csv"))
df_downregulation = pd.read_csv(os.path.join(data_dir, "downregulated_data_Dacinostat.csv"))
```

```python
df_upregulated.sort_values(["pvals_adj"]).head(n=10)
```

```python
df_downregulated.sort_values(["pvals_adj"]).head(n=10)
```

```python
df_upregulation.head()
```

```python
upregulated_genes = list(df_upregulated.sort_values("pvals_adj").names[:10])
```

```python
downregulated_genes = list(df_downregulated.sort_values("pvals_adj").names[:10])
```

```python
df_upregulation = df_upregulation[df_upregulation["logdose"]>0]
df_downregulation = df_downregulation[df_downregulation["logdose"]>0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math
import pandas as pd
from matplotlib.colors import LinearSegmentedColormap
import numpy as np
import os

# Define number of rows and columns for 5 plots per row
num_genes = len(upregulated_genes)
cols = 5
rows = math.ceil(num_genes / cols)  # Calculate required number of rows

# Dimensions for each subplot in inches
subplot_width = 10.6 / 25.4  # Convert mm to inches
subplot_height = 16.5 / 25.4  # Convert mm to inches

# Calculate total figure size
fig_width = subplot_width * cols * 5
fig_height = subplot_height * rows * 5

fig, axes = plt.subplots(rows, cols, figsize=(fig_width, fig_height))

# Flatten axes array in case it's a multi-dimensional array
axes = axes.flatten()

# Custom palettes
pred_palette = LinearSegmentedColormap.from_list('pred_palette', ['#FFC1E3', '#B12F8C'])
non_pred_palette = LinearSegmentedColormap.from_list('non_pred_palette', ['#A7D1E2', '#3F8AA6'])

# Create a lineplot for each gene
for i, gene in enumerate(upregulated_genes):
    # Filter and select exactly 4 data points per logdose for "pred"
    pred_data = (
        df_upregulation[df_upregulation['mode'] == 'pred']
        .groupby('logdose')
        .apply(lambda x: x.sample(4, random_state=42, replace=False))
        .reset_index(drop=True)
    )

    # Select up to 4 data points for other modes, if available
    non_pred_data = (
        df_upregulation[df_upregulation['mode'] != 'pred']
        .groupby('logdose')
        .apply(lambda x: x.sample(4, random_state=42, replace=False))
        .reset_index(drop=True)
    )

    # Combine both selections
    selected_data = pd.concat([pred_data, non_pred_data])

    # Add a unique identifier for each sample based on the groupby sampling
    selected_data['sample_id'] = (
        selected_data.groupby(['logdose', 'mode']).cumcount() + 1
    )

    # Separate data for coloring
    pred_points = selected_data[selected_data['mode'] == 'pred']
    non_pred_points = selected_data[selected_data['mode'] != 'pred']

    # Generate distinct colors for each group
    pred_colors = [pred_palette(x) for x in np.linspace(0, 1, len(pred_points['sample_id'].unique()))]
    non_pred_colors = ["#3F8AA6"] * 4

    # Scatter plot for 'pred' data points
    sns.scatterplot(
        data=pred_points,
        x='logdose',
        y=gene,
        hue='sample_id',  # Color by sample ID
        palette=pred_colors,
        s=50,
        ax=axes[i],
        legend=False,  # Disable legend here
    )

    # Scatter plot for 'non-pred' data points
    sns.scatterplot(
        data=non_pred_points,
        x='logdose',
        y=gene,
        hue='sample_id',  # Color by sample ID
        palette=non_pred_colors,
        s=50,
        ax=axes[i],
        legend=False,  # Disable legend here
        edgecolor=None,
    )

    # Set y-axis range and add horizontal grid lines
    axes[i].set_ylim(-1.2, 3)
    axes[i].set_xlim(0.8, 4.2)
    axes[i].yaxis.set_major_locator(plt.MultipleLocator(1))  # Grid at each integer y-tick
    axes[i].xaxis.set_major_locator(plt.MultipleLocator(1))  # Grid at each integer x-tick
    axes[i].grid(which='major', linestyle='solid', linewidth=0.5, color='gray', alpha=0.7)

    # Title
    axes[i].set_title(f'{gene}', size=10)
    axes[i].set_ylim((-0.5, 3.7))

# Remove any unused subplots if the number of genes isn't a multiple of 5
for i in range(num_genes, rows * cols):
    fig.delaxes(axes[i])

fig.savefig(os.path.join(out_dir, "upregulated_genes_cell.pdf"))
plt.tight_layout()
plt.show()
```
