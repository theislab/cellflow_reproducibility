```r
library(tidyverse)
library(glue)
```

```r
color_dict = c(
    "CellFlow"= "#B12F8C",
    "chemCPA"= "#A2E5E7",
    "Biolord"= "#BD6DED",
    "Baseline"="#E1C8AC",
    "CondOT"="#DEA1A6")
```

```r
output_dir <- glue("/lustre/groups/ml01/workspace/ot_perturbation/figures/sciplex/benchmark/plots_cf_reconstruction")
dir.create(output_dir)
```

```r
output_dir
```

# Decoded

```r
results_decoded <- read_csv("/lustre/groups/ml01/workspace/ot_perturbation/data/sciplex/results/allocation/recon_cf_metrics_decoded.csv") 
results_decoded <- results_decoded %>%
  filter(dose == split_to_filter)

```

```r
results_decoded
```

```r
results_decoded$model <- factor(results_decoded$model, 
                                levels = c("id", "condot", "chemCPA", "biolord", "otfm"),
                                labels = c("Baseline", "CondOT", "chemCPA", "Biolord", "CellFlow"))  # Correct factorization

p <- ggplot(results_decoded, aes(x=metric, y=value, fill=model)) +
  geom_boxplot(outlier.size=2) +
  labs(x="Model", y="R squared normalized gene expression") +
  scale_fill_manual(values=color_dict) +  # Removed labels argument
  scale_x_discrete(labels=c("")) +  # Keep this if you want no x-axis labels
  theme_bw() +
  theme(
    axis.title.x = element_text(size=16, face="bold"),
    axis.title.y = element_text(size=16, face="bold"),
    axis.text.x = element_text(size=14),
    axis.text.y = element_text(size=14),
    legend.title = element_text(size=14),
    legend.text = element_text(size=12),
    legend.position = "right",
    legend.box.background = element_rect(color="black"),
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank()
  ) +
  ylim(0, NA)  # Ensure the y-axis starts at 0
p
```

```r
ggsave(file.path(output_dir, "test_r_squared_decoded.pdf"), plot = p, device = "pdf")
```
