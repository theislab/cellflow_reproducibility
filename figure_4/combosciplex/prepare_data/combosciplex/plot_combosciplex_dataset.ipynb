```python
import anndata
import scanpy as sc
import os
sc.set_figure_params(scanpy=True, dpi=80, dpi_save=400)

```

```python
fig_dir = "/lustre/groups/ml01/workspace/ot_perturbation/figures/combosciplex/data"
```

```python
split = 1
```

```python
adata_train_path = f"/lustre/groups/ml01/workspace/ot_perturbation/data/combosciplex/adata_train_{split}.h5ad"
adata_test_path = f"/lustre/groups/ml01/workspace/ot_perturbation/data/combosciplex/adata_test_{split}.h5ad"
adata_ood_path = f"/lustre/groups/ml01/workspace/ot_perturbation/data/combosciplex/adata_ood_{split}.h5ad"
```

```python
adata_train = sc.read(adata_train_path)
adata_test = sc.read(adata_test_path)
adata_ood = sc.read(adata_ood_path)
```

```python
adata = anndata.concat((adata_train, adata_test, adata_ood[adata_ood.obs["control"]==0]))
```

```python
combosciplex_color_dict = {
    'Alvespimycin+Pirarubicin': '#f2d7d5',
    'Cediranib+PCI-34051': '#cd6155',
    'Dacinostat+Danusertib': '#641e16',
    'Dacinostat+Dasatinib': '#ebdef0',
    'Dacinostat+PCI-34051': '#76448a',
    'Givinostat+Carmofur': '#8e3101',
    'Givinostat+Cediranib': '#d4e6f1',
    'Givinostat+Crizotinib': '#2471a3',
    'Givinostat+Curcumin': '#d1f2eb',
    'Givinostat+Dasatinib': '#48c9b0',
    'Givinostat+SRT1720': '#EB984E',
    'Givinostat+SRT2104': '#16a085',
    'Givinostat+Sorafenib': '#fcf3cf',
    'Givinostat+Tanespimycin': '#b7950b',
    'Panobinostat+Alvespimycin': '#fad7a0',
    'Panobinostat+Crizotinib': '#8E1382',
    'Panobinostat+Curcumin': '#f8c471',
    'Panobinostat+Dasatinib': '#ce606c',
    'Panobinostat+PCI-34051': '#795548',
    'Panobinostat+SRT1720': '#22bb34',
    'Panobinostat+SRT2104': '#8D6E63',
    'Panobinostat+SRT3025': '#9d8303',
    'Panobinostat+Sorafenib': '#d7fdfd',
    'SRT2104+Alvespimycin': '#7e5109',
    'SRT3025+Cediranib': '#f4d03f',
    'control+Alvespimycin': '#b7950b',
    'control+Dacinostat': '#d6a5fa',
    'control+Dasatinib': '#2d62a3',
    'control+Givinostat': '#117864',
    'control+Panobinostat': '#b1cd3f',
    'control+SRT2104': '#af601a',
    'control': '#000000',
    'not ood': '#B6B6B6',
}
```

```python
order = list(adata[adata.obs["ood_1"]=="not ood"].obs_names) + list(adata[adata.obs["ood_1"]!="not ood"].obs_names) 
sizes = adata[order].obs.apply(lambda x: 2.0 if x["ood_1"]=="not ood" else 10.0, axis=1)
fig = sc.pl.umap(adata[order], color='ood_1', palette=combosciplex_color_dict, s=sizes.values, return_fig=True)
fig.savefig(os.path.join(fig_dir, "split_1.pdf"))
```

```python
order = list(adata[adata.obs["control"]==0].obs_names) + list(adata[adata.obs["control"]==1].obs_names) 
sizes = adata[order].obs.apply(lambda x: 2.0 if not x["control"] else 10.0, axis=1)
fig = sc.pl.umap(adata[order], color="control_str", palette=combosciplex_color_dict, s=sizes.values, return_fig=True)
fig.savefig(os.path.join(fig_dir, "control_cells.pdf"))
```
