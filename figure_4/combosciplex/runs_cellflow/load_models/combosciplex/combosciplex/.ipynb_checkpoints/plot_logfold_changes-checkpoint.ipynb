```r
library(tidyverse)
```

```r
log_fold_changes <- read_csv("/lustre/groups/ml01/workspace/ot_perturbation/analysis/combosciplex/logfoldchanges_split_3.csv") 
```

```r
output_dir <- "/lustre/groups/ml01/workspace/ot_perturbation/figures/combosciplex/logfoldchanges"
```

```r
unique_vals <- unique(log_fold_changes$group_x)
```

```r
df1 <- log_fold_changes %>%
  filter(group_x == unique_vals[[1]])

model <- lm(logfoldchanges_x ~ logfoldchanges_y, data = df1)
r_squared <- summary(model)$r.squared

p <- ggplot(df1, aes(x = logfoldchanges_x, y = logfoldchanges_y)) +
  geom_point(size = 5, alpha = 0.7, color="#3F8AA6") +  # Adjusting point size and adding transparency
  scale_color_manual(values = c("TRUE" = "blue", "FALSE" = "gray")) +
  theme_minimal(base_size = 14) +  
  geom_abline(slope = 1, intercept = 0, color="black", size=1) +
  coord_cartesian(xlim=c(-11, 11), ylim = c(-11, 11)) + 
  theme(
    panel.grid.major = element_line(color = "lightgray", size = 1.0),  # Making grid lines subtler
    panel.grid.minor = element_blank(),  # Removing minor grid lines for a cleaner look
    axis.line = element_line(color = "black"),  # Adding axis lines
    legend.position = "top",  # Placing the legend at the top for better visibility
    legend.title = element_text(size = 12),  # Adjusting legend title size
    legend.text = element_text(size = 10),    # Adjusting legend text size
    plot.margin = unit(c(1, 1, 4, 1), "lines")  # Adding extra space at the bottom for annotation
  ) 
ggsave(file.path(output_dir, "logfoldchanges_1.png"), plot = p, device = "pdf", width = 5, height = 5)  # Adjust height to accommodate the annotation
p
print(r_squared)
print(unique_vals[[1]])
```
