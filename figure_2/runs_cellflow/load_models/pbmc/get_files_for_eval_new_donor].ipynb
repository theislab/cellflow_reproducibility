```python
import os
from datetime import datetime
cutoff_date = datetime(2025, 1, 19)

# Get all files created after the cutoff date
files_after_cutoff = []
for file_name in os.listdir(out_dir):
    file_path = os.path.join(out_dir, file_name)
    if os.path.isfile(file_path):
        creation_time = datetime.fromtimestamp(os.path.getctime(file_path))
        if creation_time > cutoff_date:
            files_after_cutoff.append(file_name)

# Print the result
print("Files created after January 19th:")
for file in files_after_cutoff:
    print(file)
```
