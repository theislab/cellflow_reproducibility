```r
library(dplyr)
library(ggplot2)
library(tidyverse)
# Install glue if not already installed
if (!requireNamespace("glue", quietly = TRUE)) {
  install.packages("glue")
}
library(glue)
```

```r
data_dir = "/lustre/groups/ml01/workspace/cell_flow_zebrafish/results/single_condition/cell_type_proportion_downstream"
output_dir <- "/lustre/groups/ml01/workspace/ot_perturbation/figures/zebrafish/logfoldchanges"
```

```r
df1 <- read_csv(file.path(data_dir, glue("notochord_between_conditions.csv")))


p <- ggplot(df1, aes(x = logratio_true, y = logratio_pred)) +
  geom_point(size = 5, alpha = 0.7, color="#3F8AA6") +  # Adjusting point size and adding transparency
  scale_color_manual(values = c("TRUE" = "blue", "FALSE" = "gray")) +
  theme_minimal(base_size = 14) +  
  geom_abline(slope = 1, intercept = 0, color="black", size=1) +
  coord_cartesian(xlim=c(-8, 1), ylim = c(-8, 1)) + 
  theme(
    panel.grid.major = element_line(color = "lightgray", size = 1.0),  # Making grid lines subtler
    panel.grid.minor = element_blank(),  # Removing minor grid lines for a cleaner look
    axis.line = element_line(color = "black"),  # Adding axis lines
    legend.position = "top",  # Placing the legend at the top for better visibility
    legend.title = element_text(size = 12),  # Adjusting legend title size
    legend.text = element_text(size = 10),    # Adjusting legend text size
    plot.margin = unit(c(1, 1, 4, 1), "lines")  # Adding extra space at the bottom for annotation
  ) 
ggsave(file.path(output_dir, "logfoldchanges_notochord.png"), plot = p, device = "pdf", width = 5, height = 5)  # Adjust height to accommodate the annotation
p
```
