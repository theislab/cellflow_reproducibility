```r
library(tidyverse)
library(glue)
```

```r
color_dict = c(
    "CellFlow"= "#B12F8C",
    "Mean model 1"= "#d8dfa5",
    "Mean model 2"= "#88afd7",
    "Identity"="#E1C8AC",
    "min"="#475a69")
```

```r
output_dir <- glue("/lustre/groups/ml01/workspace/ot_perturbation/figures/zebrafish/benchmark/full_gene")
dir.create(output_dir)
```

```r
results <- read_csv("/lustre/groups/ml01/workspace/ot_perturbation/data/zebrafish_new/results/allocation/results_allocated_full_gene.csv")
```

```r
results$model <- factor(results$model, 
                                levels = c("id", "same_timepoint", "same_perturbation", "cf"),
                                labels = c("Identity", "Mean model 1", "Mean model 2", "CellFlow"))  # Correct factorization
```

```r

p <- ggplot(results, aes(x=model, y=ood_mmd, fill=model)) +
  geom_boxplot(outlier.size=2) +
  labs(x="Model", y="MMD") +
  scale_fill_manual(values=color_dict) +  # Removed labels argument
  scale_x_discrete(labels=c("")) +  # Keep this if you want no x-axis labels
  theme_bw() +
  theme(
    axis.title.x = element_text(size=16, face="bold"),
    axis.title.y = element_text(size=16, face="bold"),
    axis.text.x = element_text(size=14),
    axis.text.y = element_text(size=14),
    legend.title = element_text(size=14),
    legend.text = element_text(size=12),
    legend.position = "right",
    legend.box.background = element_rect(color="black"),
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank()
  )
p
```

```r
ggsave(file.path(output_dir, "mmd.pdf"), plot = p, device = "pdf")
```
