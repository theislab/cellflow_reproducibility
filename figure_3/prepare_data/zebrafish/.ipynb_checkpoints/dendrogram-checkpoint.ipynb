# Load here again

```python
import scanpy as sc
import os
fig_dir = "/lustre/groups/ml01/workspace/ot_perturbation/figures/zebrafish/data"
```

```python
adata = sc.read_h5ad("/lustre/groups/ml01/workspace/ot_perturbation/data/zebrafish_new/zebrafish_processed_with_umap.h5ad")
```

```python
fig = sc.pl.umap(adata, color="timepoint", return_fig=True)
```

```python
fig.savefig(os.path.join(fig_dir, "umap_temporal_coloring.png"), dpi=300)
```

```python
palette = {"False": "#B6B6B6", "True": "#000000"}
adata_reordered = adata[list(adata[~adata.obs["first_t_control"]].obs_names)+list(adata[adata.obs["first_t_control"]].obs_names)]
sc.pl.umap(adata_reordered, color="first_t_control", palette=palette, groups=["True"], size=5)
```

```python
fig.savefig(os.path.join(fig_dir, "umap_first_t_control.png"), dpi=300)
```

```python
palette = {"False": "#B6B6B6", "True": "#000000"}
adata_reordered = adata[list(adata[~adata.obs["is_control"]].obs_names)+list(adata[adata.obs["is_control"]].obs_names)]
fig = sc.pl.umap(adata_reordered, color="is_control", palette=palette, groups=["True"], return_fig=True)
```

```python
fig.savefig(os.path.join(fig_dir, "umap_is_control.png"), dpi=300)
```

```python
fig = sc.pl.umap(adata, color="tissue", return_fig=True)
```

```python
fig.savefig(os.path.join(fig_dir, "tissue.png"), dpi=300)
```

```python
fig = sc.pl.umap(adata, color="cell_type_broad", return_fig=True)
```

```python
fig.savefig(os.path.join(fig_dir, "cell_type_broad.png"), dpi=300)
```
